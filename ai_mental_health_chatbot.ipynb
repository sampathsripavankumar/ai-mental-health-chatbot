```python
from transformers import pipeline
sentiment_model = pipeline(
    "text-classification",
    model="j-hartmann/emotion-english-distilroberta-base",
    return_all_scores=False
)

print(" Model loaded successfully!")
```

```python
responses = {
    "joy":      "😊 That's wonderful! I'm happy to hear you're feeling great. Keep that positive energy!",
    "sadness":  "💙 I'm sorry you're feeling sad. It's okay to feel this way. Don,t worry about and look forward about your future.",
    "anger":    "😤 I understand you're frustrated. Take a deep breath. Things will get better.",
    "fear":     "🤗 It's okay to feel scared sometimes. You're stronger than you think!",
    "disgust":  "😟 That sounds really unpleasant. I'm here to listen if you want to share more.",
    "surprise": "😮 Wow, sounds unexpected! How are you feeling about it?",
    "neutral":  "🙂 I hear you. Tell me more about how your day is going?"
}

def mental_health_chatbot(user_message):
    if not user_message.strip():
        return "Please type something so I can help you! 😊"

    # Detect emotion
    result = sentiment_model(user_message)[0]
    emotion = result["label"].lower()
    confidence = round(result["score"] * 100, 2)

    # Get response
    bot_response = responses.get(emotion, "I'm here for you. Tell me more.")

    # Final output
    output = f"🧠 Detected Emotion: **{emotion.upper()}** ({confidence}% confidence)\n\n"
    output += f"🤖 Chatbot: {bot_response}"

    return output

print(" Chatbot logic ready!")
```

```python
import gradio as gr
interface = gr.Interface(
    fn=mental_health_chatbot,
    inputs=gr.Textbox(
        lines=3,
        placeholder="How are you feeling today? Type here...",
        label="Your Message"
    ),
    outputs=gr.Markdown(label="AI Response"),
    title="AI Mental Health Chatbot",
    description="This chatbot detects your emotion using AI and responds with empathy. Built by Sampath | AIML Final Year Project",
    examples=[
        ["I am feeling very happy today!"],
        ["I lost my job and I feel terrible"],
        ["I am so angry at everything"],
        ["I am scared about my future"]
    ],
    theme=gr.themes.Soft()
)

interface.launch(share=True)  # share=True gives a public link!
```
